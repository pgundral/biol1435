```python
# Import packages.
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Set a random seed for reproducibility.
np.random.seed(42)
# Simulate 1000 Bernoulli trials (a Binomial distribution with n=1 trial per experiment).
# Note: p = 0.5 (fair coin flip), where 0 = "heads" and 1 = "tails".
coin_flips = np.random.binomial(n=1, p=0.5, size=1000)
# Count the occurrences of 0s ("heads") and 1s ("tails").
n_heads_tails = np.array([np.sum(coin_flips == 0), np.sum(coin_flips == 1)])
# Print the results.
print(f'Heads = {n_heads_tails[0]}; Tails = {n_heads_tails[1]}.')
# Show the results array.
n_heads_tails
```

```python
# Set figure resolution.
plt.figure(figsize=(2, 1), dpi=300) # 2 inches wide, 1 inches tall.
# Create a pie chart with labels and percentages.
plt.pie(
    n_heads_tails, # Simulated results.
    labels=['Heads (0)', 'Tails (1)'],  # Clarify mapping of 0/1 to labels.
    autopct='%1.1f%%', # Round percentages to one decimal place.
    startangle=90,  # Vertical orientation.
    colors=['#E69F00', "#66b3ff"],  # Custom colors, where orange = "heads" and sky blue = "tails".
    textprops={'fontsize': 6}, # Change the fontsize of labels and percentages.
)
# Add a title and equal aspect ratio to ensure the pie chart is circular.
plt.title(r'Simulated Coin Flips ($n=1000$)', fontsize=8)
plt.axis('equal')
# Display the plot.
plt.show()
```
